```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV files
graph_df = pd.read_csv('deerhunterGraphTest.csv')
tree_df = pd.read_csv('deerhunterTreeTest.csv')

# Correct the column name in the tree DataFrame
tree_df.rename(columns={"Visided Nodes": "Visited Nodes"}, inplace=True)

# Plot Nodes Visited vs Depth Limit
plt.figure(figsize=(12, 6))
plt.plot(graph_df["Depth Limit"], graph_df["Visited Nodes"], label="Graph-Based Approach", marker="o")
plt.plot(tree_df["Depth Limit"], tree_df["Visited Nodes"], label="Tree-Based Approach", marker="o")
plt.xlabel("Depth Limit")
plt.ylabel("Visited Nodes")
plt.title("Deer Hunter: Nodes Visited vs Depth Limit (Graph vs Tree)")
plt.legend()
plt.grid(True)
plt.show()

# Plot Time Taken (ms) vs Depth Limit
plt.figure(figsize=(12, 6))
plt.plot(graph_df["Depth Limit"], graph_df["Time Taken (ms)"], label="Graph-Based Approach", marker="o")
plt.plot(tree_df["Depth Limit"], tree_df["Time Taken (ms)"], label="Tree-Based Approach", marker="o")
plt.xlabel("Depth Limit")
plt.ylabel("Time Taken (ms)")
plt.title("Deer Hunter: Time Taken (ms) vs Depth Limit (Graph vs Tree)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import numpy as np
from scipy.interpolate import make_interp_spline
import matplotlib.pyplot as plt
import csv

# Dataset
x = []
y = []
x1 = []
y1 = []
with open('graphTestRaiders.csv','r') as csvfile:
    lines = csv.reader(csvfile, delimiter=',')
    next(lines)  # Skip the first row (header)
    for row in lines:
        x.append(float(row[0]))
        y.append(float(row[1]))
X_Y_Spline = make_interp_spline(x, y)        
with open('treeTestRaiders.csv','r') as csvfile:
    lines1 = csv.reader(csvfile, delimiter=',')
    next(lines1)  # Skip the first row (header)
    for row1 in lines1:
        x1.append(float(row1[0]))
        y1.append(float(row1[1]))
X1_Y1_Spline = make_interp_spline(x1, y1)

# Returns evenly spaced numbers
# over a specified interval.
X_ = np.linspace(min(x), max(x), 500)
Y_ = X_Y_Spline(X_)
X1_ = np.linspace(min(x1), max(x1), 500)
Y1_ = X1_Y1_Spline(X1_)
# Plotting the Graph
plt.plot(X_, Y_)
plt.plot(X1_, Y1_)
plt.title("Nodes Visited for Tree Based (Orange) vs Graph Based (Blue) Solution on Raiders")
plt.xlabel("Depth")
plt.ylabel("Visited Nodes")
plt.show()



```

```python
import numpy as np
import numpy as np
from scipy.interpolate import make_interp_spline
import matplotlib.pyplot as plt
import csv

# Dataset
x = []
y = []
x1 = []
y1 = []
with open('graphTestRaiders.csv','r') as csvfile:
    lines = csv.reader(csvfile, delimiter=',')
    next(lines)  # Skip the first row (header)
    for row in lines:
        x.append(float(row[0]))
        y.append(float(row[2]))

X_Y_Spline = make_interp_spline(x, y)
with open('treeTestRaiders.csv','r') as csvfile:
    lines1 = csv.reader(csvfile, delimiter=',')
    next(lines1)  # Skip the first row (header)
    for row1 in lines1:
        x1.append(float(row1[0]))
        y1.append(float(row1[2]))
X1_Y1_Spline = make_interp_spline(x1, y1)

# Returns evenly spaced numbers
# over a specified interval.
X_ = np.linspace(min(x), max(x), 500)
Y_ = X_Y_Spline(X_)
X1_ = np.linspace(min(x1), max(x1), 500)
Y1_ = X1_Y1_Spline(X1_)


# Plotting the Graph
plt.plot(X_, Y_)
plt.plot(X1_, Y1_)
plt.title("Time Taken for Tree Based (Orange) vs Graph Based (Blue) Solution on Raiders")
plt.xlabel("Depth")
plt.ylabel("Time (ms)")
plt.show()
```
