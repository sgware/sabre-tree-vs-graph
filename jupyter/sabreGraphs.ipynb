```python
import numpy as np
import numpy as np
from scipy.interpolate import make_interp_spline
import matplotlib.pyplot as plt
import csv

# Dataset
x = []
y = []
x1 = []
y1 = []
with open('graphTestRaiders.csv','r') as csvfile:
    lines = csv.reader(csvfile, delimiter=',')
    next(lines)  # Skip the first row (header)
    for row in lines:
        x.append(float(row[0]))
        y.append(float(row[2]))

X_Y_Spline = make_interp_spline(x, y)
with open('treeTestRaiders.csv','r') as csvfile:
    lines1 = csv.reader(csvfile, delimiter=',')
    next(lines1)  # Skip the first row (header)
    for row1 in lines1:
        x1.append(float(row1[0]))
        y1.append(float(row1[2]))
X1_Y1_Spline = make_interp_spline(x1, y1)

# Returns evenly spaced numbers
# over a specified interval.
X_ = np.linspace(min(x), max(x), 500)
Y_ = X_Y_Spline(X_)
X1_ = np.linspace(min(x1), max(x1), 500)
Y1_ = X1_Y1_Spline(X1_)


# Plotting the Graph
plt.plot(X_, Y_)
plt.plot(X1_, Y1_)
plt.title("Time Taken for Tree Based (Orange) vs Graph Based (Blue) Solution on Raiders")
plt.xlabel("Depth")
plt.ylabel("Time (ms)")
plt.show()
```
